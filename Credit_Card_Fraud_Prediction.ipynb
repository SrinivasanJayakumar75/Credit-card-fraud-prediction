Applying PCA to the data

```python
import pandas as pd
from sklearn.decomposition import PCA

# Select the features you want to include in the PCA
selected_features = df.iloc[:, :32]

# Standardize the data
#standardized_features = (selected_features - selected_features.mean()) / selected_features.std()

# Apply PCA
pca = PCA(n_components=15)
principal_components = pca.fit_transform(selected_features)

# Get the loadings for the first principal component
loadings = pca.components_[0]

# Create a DataFrame with the feature names and their loadings
loading_df = pd.DataFrame({'Feature': selected_features.columns, 'Loading': loadings})

# Sort the DataFrame by absolute loading values
loading_df['Abs_Loading'] = loading_df['Loading'].abs()
loading_df = loading_df.sort_values(by='Abs_Loading', ascending=False)

# Display the DataFrame
print(loading_df)
```

```python
X = df.iloc[:,:-1]
```

```python
Y = df.iloc[:,-1]
```

Splitting the data into train and test

```python
# spliting the data into train test
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(X, Y, random_state = 1, test_size = 0.3)
```

Applying smote on imbalanced Data

```python
#Handling imbalance data
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy = 1, k_neighbors = 2)

over_X, over_Y = sm.fit_resample(X, Y)
```

```python
over_X.shape
```

```python
over_Y.shape
```

```python
from collections import Counter
Counter(over_Y)
```

Model Training

```python
#Model building
#Training the model in multiple algorithms
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(over_X, over_Y)
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(y_test, model.predict(x_test))
a
```

```python
#Random Forest Model
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()
model1.fit(over_X, over_Y)
```

```python
from sklearn.metrics import accuracy_score
b = accuracy_score(y_test, model1.predict(x_test))
b
```

```python
#Decisiontreeclassifier
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(over_X, over_Y)
```

```python
from sklearn.metrics import accuracy_score
c = accuracy_score(y_test, model2.predict(x_test))
c
```

```python
#Support vector machines
from sklearn.svm import LinearSVC
model3 = LinearSVC()
model3.fit(over_X, over_Y)
```

```python
from sklearn.metrics import accuracy_score
d = accuracy_score(y_test, model3.predict(x_test))
d
```

Comparing the model accuracy scores

```python
acc_smote = [['SVC Classifier', d], ['RandomForestClassifier', b],
             ['DecisionTreeClassifier', c], ['LogisticRegression', d]]

credit_card_fraud_prediction = pd.DataFrame(acc_smote, columns = ['Classification_models', 'accuracy_score'])
credit_card_fraud_prediction
```
